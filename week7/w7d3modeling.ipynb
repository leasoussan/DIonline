```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import sklearn as sk 
from sklearn import preprocessing
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
```

```python
df =titanic
```

```python
df
```

```python
df = df.drop(columns=['Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'])
```

```python
df =df.set_index('PassengerId')
```

```python
df =pd.get_dummies(df, columns =['Pclass'], prefix= 'Class')
```

```python
df['Sex'] = [1 if sex == 'male' else 0 for sex in df['Sex']] 
```

```python
df
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
# we would like to seperate X from Y 
Y =  df['Survived']
X = df.drop(columns=['Survived'])
```

```python
ms_scaler = preprocessing.MinMaxScaler()
X_minmax = ms_scaler.fit_transform(X)
X_minmax = pd.DataFrame(data = X_minmax, columns= X.columns)
```

```python
X_minmax
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
lr =LinearRegression()
knn=KNeighborsClassifier()
gnb= GaussianNB()
svm = SVC()
```

```python
lr.fit(X_train, y_train)
```

```python
lr_prediction = lr.predict(X_test)
lr_prediction  = [1 if num >= 0.5 else 0 for num in lr_prediction]
```

```python
knn.fit(X_train, y_train)
gnb.fit(X_train, y_train)
svm.fit(X_train, y_train)
```

```python
prdc_knn = knn.predict(X_test)
prdc_gnb = gnb.predict(X_test)
prdc_svm = svm.predict(X_test)
```

```python
prediction = pd.DataFrame({"LR": lr_prediction, "KNN": prdc_knn,"GNB":prdc_gnb, "SVM": prdc_svm, "Real": y_test})
```
