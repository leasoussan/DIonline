```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')
```

```python
cars.head(10)
```

```python
corr = cars.corr()
```

```python
sns.heatmap(corr)
```

```python
df=cars
```

```python
df
```

```python
new = df["Price"]/df["MPG.city"]
df.insert(4,'price/MPG', new)
```

```python
df
```

```python
plt.bar(df["price/MPG"], df["Max.Price"])
plt.xlabel("Price Meter Galon")
plt.ylabel("Max Price of car")
plt.show()
```
